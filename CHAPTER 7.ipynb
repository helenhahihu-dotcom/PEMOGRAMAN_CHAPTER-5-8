```python
import numpy as np  # mengimpor numpy untuk operasi numerik
from scipy import integrate
if not hasattr(integrate, "trapz"): # menggunakan alias karena scipy yang digunakan sekarang adalah ver. baru
    integrate.trapz = integrate.trapezoid
if not hasattr(integrate, "simps"):
    integrate.simps = integrate.simpson

def integrate_rec(f, a, b, n):  # mendefinisikan fungsi integrasi metode persegi panjang
    # Implementation of the rectanlge method 
    h = (b-a)/n  # menghitung lebar tiap subinterval
    x = np.linspace(a, b, n+1)  # membuat titik pembagi interval
    i=0  # inisialisasi indeks perulangan
    area=0  # inisialisasi luas total
    while i<n:  # melakukan perulangan sebanyak n interval
        sup_rect =  f(x[i])*h  # menghitung luas persegi panjang pada titik ke-i
        area += sup_rect  # menambahkan luas ke total
        i += 1  # menaikkan indeks
    return area  # mengembalikan hasil integrasi
'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_10 = integrate_rec(np.sin, 0, np.pi, 10) # menghitung integral dengan 10 interval
sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5)  # menghitung integral dengan 5 interval
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10)  # menghitung integral dengan 10 interval

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_10))  # menampilkan hasil untuk n=5
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_5))  # menampilkan hasil untuk n=10
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_10))  # menampilkan hasil untuk n=100
     
```

```python
import numpy as np  # mengimpor library numpy dengan alias np untuk operasi numerik

def integrate_trap(f, a, b, n):  # mendefinisikan fungsi integrasi metode trapesium
    # Implementation of the trapezoidal rule
    h = (b-a)/n  # menghitung lebar tiap subinterval (step size)
    x = np.linspace(a, b, n+1)  # membuat array titik dari a sampai b sebanyak n+1 titik
    i=1  # inisialisasi indeks awal untuk perulangan
    area = h*(f(x[0]) + f(x[n]))/2  # menghitung luas awal dari dua ujung (rumus trapesium)
    while i<n:  # perulangan untuk menjumlahkan luas bagian tengah
        sup_rect = f(x[i])*h  # luas persegi panjang pada titik ke-i
        area += sup_rect  # menambahkan ke total luas
        i += 1  # menaikkan indeks
    return area  # mengembalikan hasil aproksimasi integral

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5)  # menghitung integral sin(x) dengan 5 partisi
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10)  # sama seperti di atas (tidak dijelaskan ulang)

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5))  # menampilkan hasil dengan format 2 angka desimal
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10))  # sama seperti di atas

'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
'''
```

```python
import numpy as np  # import numpy untuk perhitungan numerik
from scipy import integrate  # import fungsi integrasi dari scipy


x = np.linspace(0,9, 3)  # membuat 3 titik pembagi (n=2 interval)
y = x**2  # fungsi y = x kuadrat

sup_trapz = integrate.trapz(y,x)  # integral dengan metode trapesium ver. lama
sup_simps = integrate.simps(y,x)  # integral dengan metode Simpson


print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz))  # hasil trapesium
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps))  # hasil Simpson

'''
Output:
Using n=2, the trapezoidal rule returns a value of 273
Using n=2, the composite Simpson rule returns a value of 243
'''
```

```python
import numpy as np  # mengimpor NumPy untuk membuat array dan operasi numerik
from scipy import integrate  # mengimpor modul integrasi dari SciPy
# membuat alias agar kompatibel dengan SciPy versi baru
if not hasattr(integrate, "trapz"):  # cek apakah fungsi trapz tersedia
    integrate.trapz = integrate.trapezoid  # jika tidak ada, gunakan trapezoid sebagai pengganti
if not hasattr(integrate, "simps"):  # cek apakah fungsi simps tersedia
    integrate.simps = integrate.simpson  # jika tidak ada, gunakan simpson sebagai pengganti

conturs_areas = np.array([200000, 150000, 90000, 50000, 25000, 12000, 6000])  # array berisi luas penampang (m^2) pada tiap titik
x = np.array([0, 30, 60, 90, 120, 150, 180]) # array jarak antar penampang (meter)
vol_traps = integrate.trapz(conturs_areas, x) # menghitung volume menggunakan metode trapesium
vol_simps = integrate.simps(conturs_areas, x) # menghitung volume menggunakan metode Simpson (lebih akurat)
print('Trapezoidal volume: {:.0f} m^3'.format(vol_traps)) # menampilkan hasil volume metode trapesium tanpa desimal
print('Simpson volume: {:.0f} m^3'.format(vol_simps)) # menampilkan hasil volume metode Simpson tanpa desimal
```

```python
def simple_lithopress(z,ro=2600, g=9.81):  # fungsi tekanan litostatik sederhana
    pressure = z*g*ro/1e6  # menghitung tekanan dalam MPa
    return pressure  # mengembalikan nilai tekanan

my_pressure = simple_lithopress(z=2000)  # menghitung tekanan pada kedalaman 2000 m
print('The pressure at 2000 meters is {0:.0f} MPa'.format(my_pressure))  # menampilkan hasil
```

```python
import numpy as np  # import numpy untuk array dan linspace
from scipy.integrate import trapz  # import trapz untuk integrasi (tidak dipakai langsung di plot)
import matplotlib.pyplot as plt  # import matplotlib untuk plotting

r = np.linspace(1,6400,6400)  # membuat array jari-jari bumi dari pusat ke permukaan

def density():  # fungsi model densitas bumi (versi dimodifikasi)
    ro_inner_core = np.linspace(13200, 12700, 1220)  # inti dalam (diubah sedikit)
    ro_outer_core = np.linspace(12300, 9800, 2259)  # inti luar (diubah)
    ro_lower_mantle = np.linspace(5700, 4300, 2171)  # mantel bawah
    ro_upper_mantle = np.linspace(4200, 3300, 720)  # mantel atas
    ro_crust = np.linspace(3000, 2600, 30)  # kerak bumi
    
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust))  
    # menggabungkan semua lapisan menjadi satu array
    
    return ro_final  # kembalikan profil densitas

ro = density()  # simpan hasil fungsi densitas

fig, ax = plt.subplots()  # membuat figure dan axes
ax.plot(r,ro, label="Densities on the Earth's Interior")  # plot densitas terhadap radius
ax.set_ylabel(r"Density [Kg/m$^3$]")  # label sumbu Y
ax.set_xlabel("Distance from the Earth center r [km]")  # label sumbu X
ax.legend()  # menampilkan legenda
ax.grid()  # menampilkan grid
```

```python
def gravity(r):  # fungsi menghitung percepatan gravitasi (dimodifikasi)
    g = np.zeros(len(r))  # array untuk menyimpan nilai gravitasi
    G = 6.67430e-11  # konstanta gravitasi (nilai diperbarui)
    r = r * 1000  # konversi km → meter
    
    for i in range(1, len(r)):  # loop dari pusat ke permukaan
        
        r1 = r[:i]  # radius dari pusat sampai titik i
        ro1 = ro[:i]  # densitas pada bagian tersebut
        r_current = r[i-1]  # radius saat ini
        
        y = ro1 * r1**2  # integran (ρ * r^2)
        mass_integral = integrate.trapz(y, r1)  # integrasi massa
        
        g[i] = (4 * np.pi * G / (r_current**2)) * mass_integral  
        # menghitung gravitasi berdasarkan massa dalam radius tersebut
    
    return g  # hasil array gravitasi

g = gravity(r)  # menghitung gravitasi

fig, ax = plt.subplots()  
ax.plot(r,g)  # grafik gravitasi
ax.grid()
ax.set_ylabel(r'g [m/s$^2]$')  # label sumbu y
ax.set_xlabel('Distance from the Earth center r [km]')  # label sumbu x
```

```python
def pressure(r, ro, g):  # fungsi menghitung tekanan
    p = np.zeros(len(r))  # array untuk menyimpan hasil tekanan
    r = r *1000  # konversi km ke meter
    
    for i in range(0,len(r)):  # perhitungan di setiap titik radius    
        r1 = r[i:len(r)]  # bagian radius dari titik i sampai pusat
        ro1 = ro[i:len(r)]  # densitas pada bagian tersebut
        g1 = g[i:len(r)]  # gravitasi pada bagian tersebut
        
        y = ro1 * g1  # fungsi ρ × g (untuk integral tekanan)
        p1 = trapz(y, r1)  # integrasi numerik untuk tekanan
        p[i] = p1  # simpan hasil tekanan
    
    return p  # hasil tekanan

p = pressure(r,ro,g)/1e9  # konversi tekanan ke GPa
z = np.linspace(6400, 1, 6400)  # skala kedalaman dari permukaan ke pusat

fig, ax = plt.subplots()  
ax.plot(z,p)  # plot tekanan terhadap kedalaman
ax.grid()
ax.set_ylabel('P [GPa]')  # label tekanan
ax.set_xlabel('Depth z from the Earth Surface [km]')  # label kedalaman
```
